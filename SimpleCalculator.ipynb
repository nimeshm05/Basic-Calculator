```python
from tkinter import *

def btnClick(number):
    global expression
    expression = expression + str(number)
    equation.set(expression)
def btnEquals():
    global expression
    total = str(eval(expression))
    equation.set(total)
    expression =""
def btnClear():
    global expression
    expression = ""
    equation.set("")

cal = Tk()
cal.title("A Simple Calculator")
expression = ""
equation = StringVar()
cal.resizable(width=False, height=False)
txt_display = Entry(cal, font = ('arial', 20), textvariable=equation, justify = 'right')
txt_display.grid(row = 1, column = 1, columnspan = 4)

button7 = Button(cal, fg='black', bg='white', text='7', padx=16, pady=16, font = ('arial',20,'bold'),
                command = lambda: btnClick(7))
button7.grid(row=2, column=1)

button8 = Button(cal, fg='black', bg='white', text='8', padx=16, pady=16, font = ('arial',20,'bold'),
                command = lambda:btnClick(8))
button8.grid(row=2, column=2)

button9 = Button(cal, fg='black', bg='white', text='9', padx=16, pady=16, font = ('arial',20,'bold'),
                command = lambda:btnClick(9))
button9.grid(row=2, column=3)

button4 = Button(cal, fg='black', bg='white', text='4', padx=16, pady=16, font = ('arial',20,'bold'),
                command = lambda:btnClick(4))
button4.grid(row=3, column=1)

button5 = Button(cal, fg='black', bg='white', text='5', padx=16, pady=16, font = ('arial',20,'bold'),
                command = lambda:btnClick(5))
button5.grid(row=3, column=2)

button6 = Button(cal, fg='black', bg='white', text='6', padx=16, pady=16, font = ('arial',20,'bold'),
                command = lambda:btnClick(6))
button6.grid(row=3, column=3)

button1 = Button(cal, fg='black', bg='white', text='1', padx=16, pady=16, font = ('arial',20,'bold'),
                command = lambda:btnClick(1))
button1.grid(row=4, column=1)

button2 = Button(cal, fg='black', bg='white', text='2', padx=16, pady=16, font = ('arial',20,'bold'),
                command = lambda:btnClick(2))
button2.grid(row=4, column=2)

button3 = Button(cal, fg='black', bg='white', text='3', padx=16, pady=16, font = ('arial',20,'bold'),
                command = lambda:btnClick(3))
button3.grid(row=4, column=3)
###
buttonsub = Button(cal, fg='black', bg='white', text='-', padx=16, pady=16, font = ('arial',20,'bold'),
                  command = lambda: btnClick('-'))
buttonsub.grid(row=2, column=4)

buttonadd = Button(cal, fg='black', bg='white', text='+', padx=16, pady=16, font = ('arial',20,'bold'),
                  command = lambda:btnClick("+"))
buttonadd.grid(row=3, column=4)

buttonmul = Button(cal, fg='black', bg='white', text='*', padx=16, pady=16, font = ('arial',20,'bold'),
                  command = lambda:btnClick("*"))
buttonmul.grid(row=4, column=4)

buttonclr = Button(cal, fg='black', bg='white', text='C', padx=16, pady=16, font = ('arial',20,'bold'),
                  command = lambda: btnClear())
buttonclr.grid(row=5, column=1)

button0 = Button(cal, fg='black', bg='white', text='0', padx=16, pady=16, font = ('arial',20,'bold'),
                command = lambda: btnClick(0))
button0.grid(row=5, column=2)

buttonequal = Button(cal, fg='black', bg='white', text='=', padx=16, pady=16, font = ('arial',20,'bold'),
                    command = lambda: btnEquals())
buttonequal.grid(row=5, column=3)

buttondiv = Button(cal, fg='black', bg='white', text='/', padx=16, pady=16, font = ('arial',20,'bold'),
                  command = lambda:btnClick("/"))
buttondiv.grid(row=5, column=4)

cal.mainloop()
```
